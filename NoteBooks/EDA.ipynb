```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap of Credit Card Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, hue='Class', palette='pastel', legend=False)
plt.title("Distribution of Fraudulent vs Normal Transactions (0 = Normal, 1 = Fraud)")
plt.xlabel("Transaction Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
correlation = df.corr()
top_corr = correlation['Class'].abs().sort_values(ascending=False)[1:11]  
important_features = top_corr.index.tolist()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[important_features + ['Class']].corr(), annot=True, cmap='coolwarm')
plt.title("Top Features Correlated with Class")
plt.tight_layout()
plt.show()
```

```python
plotting = ['V1', 'V2', 'V3', 'V4', 'V5']
for col in plotting:
    df[f'{col}_bin'] = pd.qcut(df[col], q=6, duplicates='drop')
    plt.figure(figsize=(10, 6))
    sns.countplot(x=f'{col}_bin', hue='Class', data=df, palette='pastel')
    plt.title(f'Fraud Count vs. {col} (binned)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
df = df.drop('id', axis = 'columns')
```

```python
df.sample()
```

```python
class_0 = df[df['Class'] == 0]['Amount'] 
class_1 = df[df['Class'] == 1]['Amount']  
```

```python
plt.figure(figsize=(12, 6))
plt.title('Transaction Amount Distribution by Fraud Class')
sns.histplot(class_0, kde=True, bins=100, color='green', label='Class 0 (Normal)', stat='density')
sns.histplot(class_1, kde=True, bins=100, color='red', label='Class 1 (Fraud)', stat='density')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```
